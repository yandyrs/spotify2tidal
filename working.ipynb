```python
import re


def sanitize_name(name):
    regex = r' \(?(feat|ft|with)\.? .+\)?'
    return re.sub(regex, '', name, flags=re.IGNORECASE).replace('(', '').replace(')', '').replace(' - ', ' ').lower()

def find_tidal_track(tidal, title, artists, album, album_artists = None):
    title = sanitize_name(title)

    # initial filter by title
    query = '{} {}'.format(title, artists[0])
    track_results = tidal.search('track', query).tracks
    results = []
    for r in track_results:
        name = sanitize_name(r.name)
        if name == title:
            results.append(r)

    # if there are no results with the initial filter, the song can't be found
    if not len(results):
        return None

    # if we can find an exact album match, prefer that
    for r in results:
        if sanitize_name(r.album.name) == sanitize_name(album):
            return r

    # otherwise, try to match an album published by the original artist
    artist_albums = tidal.get_artist_albums(results[0].artist.id) + tidal.get_artist_albums_ep_singles(results[0].artist.id) + tidal.get_artist_albums_other(results[0].artist.id)
    for a in artist_albums:
        for r in results:
            if r.album.name == a.name:
                return r

    return None
```
